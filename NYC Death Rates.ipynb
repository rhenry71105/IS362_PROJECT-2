We will begin by importing pandas and reading the .csv data. The data is captured is from [Github](https://github.com/) from Djamshed Djuraev. We will begin by importing pandas and then tidying up the data.

```python
import pandas as pd # importing pandas module for usage
import numpy as np

deaths = pd.read_csv('C:/Users/rickardo.henry/Desktop/IS362_Project2/NYC-Death-Rates.csv')
# Reading Data from CSV file.
deaths
```

<span style='color:orange'><i> We will now tidy up the data above by replacing all the <b>'.'</b> with 'NaN' values. </i></span>

```python
tidy = deaths.replace(to_replace = '.', value = np.nan)
tidy
```

<b><span style="color: #FF0000;"> Q1. Which year had the most deaths ? </span></b>

```python
deaths.groupby('Year').count()
```

<span style="color: #006600;"><b>Answer Q1. We can see 2011 had the most deaths with a total of 43.</b></span>

<span style="color: #FF0000;"><b> Q2.  Which sex had the highest death amount ?</b></span>

```python
deaths.groupby('Race Ethnicity')['Sex'].count()
```

<span style="color: #006600;"><b> Answer Q2. We can see the "Not Stated/Unknown" Race has the highest death toll with a total of 53, 
followed by "Black Non-Hispanic & Hipanics" at a total of 50.</b></span>

<span style="color: #FF0000;"><b> Q3. What were the top 5 leading causes of deaths  in 2007 ? </b></span>

```python
deaths[deaths['Year']==2007].sort_values('Deaths', ascending = False).head(5)
```
