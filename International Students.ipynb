<b><span style="color: #FF0000;"> Q1. Which country had the most international students ?</span></b>

```python
student_data.groupby('Country').count()
```

<b><span style="color: #FF0000;"> Q2.  Which country had the most female students ?</span></b>

```python
student_data.groupby('Country')['Gender'].value_counts()
```

<span style="color: #006600;"><b> Answer Q2. We can see that Australia has the most female international students. </b></span>

<b><span style="color: #FF0000;"><b> Q3. What was the most used form of commuting for students ? </b></span>
