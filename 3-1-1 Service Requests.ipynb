The data is captured from [NYC Open Data](https://opendata.cityofnewyork.us/). We will begin by importing pandas and reading the .csv data. 

```python
import pandas as pd # Importing the pandas Module For Usage.
import numpy as np

service_requests = pd.read_csv('C:/Users/rickardo.henry/Desktop/IS362_Project2/311-ServiceRequests-(200).csv')
#reading data from CSV file on Computer.
service_requests
```

<b><span style='color: #FF0000;'> Q1. What City (borough) complained the most? </span></b>

```python
borough = service_requests.groupby('Borough').count()
borough
```

<span style='color: #006600;'><b>Answer Q1. We can see that the borough of Manhattan had the most complaints with a total of 64 complaints.</b></span>

<b><span style='color: #FF0000;'> Q2.  What agency received the most complaints ?</span></b>

```python
agency = service_requests.groupby('Agency').count()
agency
```

<span style='color: #006600;'><b> Answer Q2. We can see that DEP <i>(Department of Environmetal Protection)</i> received the most complaints with a total of 148 complaints.</span></b>

<span style='color: #FF0000;'><b> Q3.  What was the most common complaint type ? </b></span>

```python
complaint = service_requests.groupby('Complaint Type').count()
complaint
```

<b><span style='color: #006600;'> Answer Q3. We can see that Lead was the complaint type most common with a total of 136 complaints.</span></b>
